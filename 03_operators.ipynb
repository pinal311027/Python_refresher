```python
#arithmatic operators
a = 5
b = 2
print("Sum = ",a + b)
print("sub = ",a - b)
print("mul = ",a * b)
print("div = ",a / b)
print("Floor div = ",a // b)
print("modules = ", a % b)
print("exponentiation = ",a ** b)
```

```python
#comparision operators
a = 8
b = 4
print("a == b :",a == b)
print("a != b :",a != b)
print("a > b :",a > b)
print("a < b :",a < b)
print("a >= b :",a >= b)
print("a <= b :",a <= b)
```

```python
#logical operators
x = True
y = False
print("both True :",x and y)
print("either True :", x or y)
print("negation :",not x )
```

```python
#bitwise operators
a = 5
b = 3
print(a & b)
print(a | b)
```

```python
#assignment operators
a = 5
print(a)
a += 5
print(a)
a -= 5
print(a)
a *= 2
print(a)
a /= 5
print(a)
a **= 2
print(a)
```

```python
#membership and identify operator
lst = [1,2,3]
x = 2
print("x in lst :",x in lst)
print("x not in lst :",x not in lst)
print("x is b :",x is b)
print("x is not b :",x is not b)
```
