```python
#list in python(mutuable,ordered)
list1= [1,2,3,4,5,6]
list1.append(7)
print(list1)
list1.insert(0,0)
print(list1)
list1.extend([8,9,10])
print(list1)
list1.remove(10)
print(list1)
list1.pop()
print(list1)
list1.index(5)
print(list1)
list1.count(7)
print(list1)
list1.sort()
print(list1)
list1.reverse()
print(list1)
new_list = list.copy()
print(new_list)
new_list.clear()
print(new_list)
```

```python
#tuple methods (unmutuable,ordered)
tup = (1,2,3,4,5,6,2,2)
print(tup.count(2))
print(tup.index(4))
print(tup[6])
print(tup[0:])
```

```python
#set method(mutuable,unordered)
empty_set = set() #empty set
set1 = {1,2,3,5,3,9,3}
set1.add(10)
print(set1)
set1.update([11,12])
print(set1)
set1.remove(12)
print(set1)
set1.discard(20)
print(set1)
set2 = set1.copy()
print(set2)
set2.clear()
print(set2)
```

```python
#set operation
set1 = {1,2,3,4,5,6}
set2 = {5,6,7,8,9,10}
print(set1.union(set2))
print(set1.intersection(set2))
print(set1.difference(set2))
```

```python
#dictionary and methods
student = {
    "name" : "pinal",
    "age" : 20,
    "grade" : "A"
}
print(student.keys())
print(student.values())
print(student.items())
print(student.get("city",0))
print(student.update({"age": 21}))
print(student.pop("grade"))
dict1 = student.copy()
print(dict1)
print(student.clear())
```
