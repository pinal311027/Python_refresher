```python
#all patterns using python
#1)square pattern
n = 4
for i in range(n):
    for j in range(n):
        print("*",end=" ")
    print()
```

```python
#2)right-angle triangle
n = 5
for i in range(n):
    for j in range(i):
        print("*",end=" ")
    print()
```

```python
#3)inverted triange
n = 5 
for i in range(n):
    for j in range(n-i):
        print("*",end=" ")
    print()
```

```python
#4)pyramid pattern
n = 5
for i in range(n):
    for j in range(n-i-1):
        print(" ",end=" ") 
    for j in range(2*i+1):
        print("*",end=" ")
    print()
```

```python
#5)inverted pyramid 
n = 5
for i in range(n):
    for j in range(i):
        print(" ",end=" ")
    for j in range(2*(n-i)-1):
        print("*",end=" ")
    print()
```

```python
#6)diamond (mix of both pyramid)
n = 5
for i in range(n):
    for j in range(n-i-1):
        print(" ",end=" ") 
    for j in range(2*i+1):
        print("*",end=" ")
    print()
for i in range(n-2,-1,-1):
    for j in range(n-i-1):
        print(" ",end=" ")
    for j in range(2*i+1):
        print("*",end=" ")
    print()
```

```python
#7)hollow square
n = 5
for i in range(n):
    for j in range(n):
        if i==0 or i== n-1 or j==0 or j == n-1:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
#8)hollow pyramid
n = 5
for i in range(n):
    for j in range(n-i-1):
        print(" ", end=" ")
    for j in range(2*i+1):
        if j==0 or j== 2*i or i == n-1:
            print("*", end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
#9)number triange
n = 5
for i in range(1,1+n):
    for j in range(1, i+1):
        print(j, end="")
    print()
```

```python
#10)repeated number triange
n = 5
for i in range(1,1+n):
    for j in range(1, i+1):
        print(i, end="")
    print()
```
