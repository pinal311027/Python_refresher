```python
#read file
file = open("example.txt","r")
content = file.read()
print(content)
file.close()
```

```python
#read first line
file = open("example.txt","r")
line1 = file.readline()
print(line1)
file.close()
```

```python
#read all line as a list
file = open("example.txt","r")
lines = file.readlines()
print(lines)
file.close()
```

```python
#write in file
file = open("examples.txt","w")
file.write("i'm 21 year old")
file.close()
```

```python
#write lines in file
lines = {"Hello\n","welcome to python\n","file handling\n"}
file = open("examples.txt","w")
file.writelines(lines)
file.close()
```

```python
#appending to a file
file = open("examples.txt","a")
file.write("\nThis is an additional line.")
file.close()
```

```python
#using with statement(best practice)
with open("example.txt","r") as file:
    content = file.read()
    print(content)
```

```python
#file exists or not
import os 
if os.path.exists("example.txt"):
    print("File exist")
else:
    print("not found..!!")
```

```python
#file delete if exists
import os 
if os.path.exists("examples.txt"):
    os.remove("examples.txt")
    print("file deleted...")
else:
    print("file does not exist")
```
