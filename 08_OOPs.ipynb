```python
#class and object
class Car:
    def __init__(self,brand,model):
        self.brand = brand
        self.model = model

    def display_info(self):
        return f"{self.brand} {self.model}"

car1 =  Car("hundai","aura")
print(car1.display_info())
```

```python
#Encapsulation
class BankAccount:
    def __init__ (self,balance):
        self.__balance = balance
    def get_balance(self):
        return self.__balance
    def deposit(self,amount):
        if amount > 0:
            self.__balance += amount

#using encapsulation
account = BankAccount(1000)
account.deposit(500)
print(account.get_balance())
```

```python
#single inheritance
class Animal :
    def spark(self):
        return"Animal makes sound"

class Dog(Animal):
    def speak (self):
        return "Bark"

dog = Dog()
print(dog.speak())
print(dog.spark())
```

```python
#multiple Inheritnce
class A:
    def method_a(self):
        return "Method A"

class B:
    def method_b(self):
        return "Method B"

class C(A,B):
    pass

obj = C()
print(obj.method_a())
print(obj.method_b())
```

```python
#polymorphism
class Bird:
    def fly(self):
        return "Birds can fly"

class Panguin (Bird):
    def fly(self):
        return "panguin can not fly"

bird = Bird()
panguin = Panguin()

print(bird.fly())
print(panguin.fly())
```

```python
#Abstraction
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Square(Shape):
    def __init__(self,side):
        self.side = side 

    def area(self):
        return self.side * self.side

square = Square(5)
print(square.area())
```

```python
#magic method
class Book:
    def __init__(self,title,pages):
        self.title = title
        self.pages = pages

    def __str__(self):
        return f"Book: {self.title}"

    def __len__(self):
        return self.pages

book = Book("Python basics", 300)
print(str(book))
print(len(book))
```
