```python
#if-elif-else statement
a = "w"
if (a in "aeiou"):
    print("a is vowel")
elif (a.isalpha()):
    print("a is constant")
else:
    print("please enter correct alphabet")
    
```

```python
#match case statement
light = input("Enter the signal colour")
match light:
    case "red":
        print("stop your car.")
    case "yellow":
        print("get ready to go.")
    case "green":
        print("Allow to goo...")
    case _:
        print("signal light is break")
```
